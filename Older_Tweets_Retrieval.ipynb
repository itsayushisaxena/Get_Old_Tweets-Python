```python
import snscrape.modules.twitter as sntwitter
```

```python
csvFile = open('15-10-2020.csv', 'a') #creates a file in which you want to store the data.
csvWriter = csv.writer(csvFile)
```

```python
maxTweets = 2550  # the number of tweets you require
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#covid19' + 'since:2020-10-15 until:2020-10-16').get_items()) :
        if i > maxTweets :
            break
        print(tweet.date)
        csvWriter.writerow([tweet.date, tweet.content.encode('utf-8')]) #If you need more information, just provide the attributes
```
